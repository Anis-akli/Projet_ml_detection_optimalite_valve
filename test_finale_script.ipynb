```python
import unittest
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
# Fonctions Ã  tester
def create_target(profile):
    return (profile[1] == 100).astype(int)

def scale_data(scaler, X_train, X_test):
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    return X_train_scaled, X_test_scaled


```

```python
# Classe de tests
class TestPreprocessing(unittest.TestCase):

    def test_target_binary(self):
        profile = pd.DataFrame({1: [100, 80, 100, 60]})
        y = create_target(profile)
        self.assertTrue(set(y.unique()).issubset({0, 1}))
        self.assertEqual(len(y), len(profile))

    def test_no_nan_in_scaled_data(self):
        X_train = pd.DataFrame({"a": [1, 2, 3], "b": [4, 5, 6]})
        X_test = pd.DataFrame({"a": [2, 3], "b": [5, 6]})
        scaler = StandardScaler()
        X_train_scaled, X_test_scaled = scale_data(scaler, X_train, X_test)
        self.assertFalse(np.isnan(X_train_scaled).any())
        self.assertFalse(np.isnan(X_test_scaled).any())

    def test_same_number_of_features(self):
        X_train = pd.DataFrame(np.random.rand(10, 4))
        X_test = pd.DataFrame(np.random.rand(5, 4))
        scaler = StandardScaler()
        X_train_scaled, X_test_scaled = scale_data(scaler, X_train, X_test)
        self.assertEqual(X_train_scaled.shape[1], X_test_scaled.shape[1])
```

```python
suite = unittest.TestLoader().loadTestsFromTestCase(TestPreprocessing)
unittest.TextTestRunner(verbosity=2).run(suite)
```
