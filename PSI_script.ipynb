### Dans ce script, nous allons analyser les fichier " PS2, PS3, PS4, PS5, PS6 " qui represente les données brutes de pression prises par les capteurs 2,3,4,5,6 .

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cycles à visualiser (toujours les mêmes pour comparaison)
cycles_exemple = [0, 50, 200]
```

```python
for i in range(2, 7):
    print(f"\n======= PS{i} ======")
    ps = pd.read_csv(f"PS{i}.txt", sep="\t", header=None)

    #Courbes temporelles 
    plt.figure(figsize=(12, 4))
    for c in cycles_exemple:
        plt.plot(ps.iloc[c], label=f"Cycle {c}")

    plt.title(f"PS{i}  Pression Cycles exemple")
    plt.xlabel("Temps")
    plt.ylabel("Pression")
    plt.legend()
    plt.show()

    #Boxplot Mean vs Median
    ps_mean = ps.mean(axis=1)
    ps_median = ps.median(axis=1)

    plt.figure(figsize=(6, 4))
    plt.boxplot([ps_mean, ps_median], labels=["Mean", "Median"])
    plt.title(f"PS{i} Mean vs Median")
    plt.ylabel("Pression")
    plt.show()

    #Distribution de l'écart-type
    ps_std = ps.std(axis=1)

    plt.figure(figsize=(6, 4))
    plt.hist(ps_std, bins=40)
    plt.title(f"PS{i}  Distribution de l'écart-type")
    plt.xlabel("STD")
    plt.ylabel("Nombre de cycles")
    plt.show()

    # Distribution des maxima
    ps_max = ps.max(axis=1)

    plt.figure(figsize=(6, 4))
    plt.hist(ps_max, bins=40)
    plt.title(f"PS{i} Distribution des maxima")
    plt.xlabel("Pression max")
    plt.ylabel("Nombre de cycles")
    plt.show()
```

```python
import pandas as pd
# VS1
vs1 = pd.read_csv("VS1.txt", sep="\t", header=None)

vs1_features = pd.DataFrame({
    "VS1_std": vs1.std(axis=1),
    "VS1_max": vs1.max(axis=1)
})

# PS1
ps1 = pd.read_csv("PS1.txt", sep="\t", header=None)

ps1_features = pd.DataFrame({
    "PS1_std": ps1.std(axis=1),
    "PS1_max": ps1.max(axis=1),
    "PS1_median": ps1.median(axis=1)
})

# PS2
ps2 = pd.read_csv("PS2.txt", sep="\t", header=None)

ps2_features = pd.DataFrame({
    "PS2_std": ps2.std(axis=1),
    "PS2_max": ps2.max(axis=1),
    "PS2_median": ps2.median(axis=1)
})

# PS3
ps3 = pd.read_csv("PS3.txt", sep="\t", header=None)

ps3_features = pd.DataFrame({
    "PS3_std": ps3.std(axis=1),
    "PS3_max": ps3.max(axis=1),
    "PS3_median": ps3.median(axis=1)
})

# PS5
ps5 = pd.read_csv("PS5.txt", sep="\t", header=None)

ps5_features = pd.DataFrame({
    "PS5_mean": ps5.mean(axis=1),
    "PS5_max": ps5.max(axis=1)
})

# PS6
ps6 = pd.read_csv("PS6.txt", sep="\t", header=None)

ps6_features = pd.DataFrame({
    "PS6_mean": ps6.mean(axis=1),
    "PS6_max": ps6.max(axis=1)
})

# MERGE FINAL
dataset_final = pd.concat([
    vs1_features,
    ps1_features,
    ps2_features,
    ps3_features,
    ps5_features,
    ps6_features
], axis=1)
dataset_final.index.name = "cycle"
dataset_final.to_csv("dataset_PSI.csv")

print("dataset_PSI.csv sauvegardé :", dataset_final.shape)

print("Shape du dataset final :", dataset_final.shape)
print(dataset_final.head())
```

```python
import pandas as pd
import numpy as np

dataset_PSI = pd.read_csv("dataset_PSI.csv", index_col=0)

psi_features = dataset_PSI.columns.tolist()

# Fonctions outliers
def outliers_zscore(series, threshold=3):
    z = (series - series.mean()) / series.std()
    return np.abs(z) > threshold

def outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return (series < Q1 - 1.5 * IQR) | (series > Q3 + 1.5 * IQR)

# Analyse
results = []

for feature in psi_features:
    z_out = outliers_zscore(dataset_PSI[feature])
    iqr_out = outliers_iqr(dataset_PSI[feature])

    results.append({
        "feature": feature,
        "Z-score_outliers_%": round(100 * z_out.mean(), 2),
        "IQR_outliers_%": round(100 * iqr_out.mean(), 2)
    })

outliers_PSI = pd.DataFrame(results)

print(outliers_PSI)
```
