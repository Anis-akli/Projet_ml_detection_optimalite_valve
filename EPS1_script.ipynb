```python
import pandas as pd
import matplotlib.pyplot as plt

# Chargement EPS1
eps = pd.read_csv("EPS1.txt", sep="\t", header=None)

print("EPS1 shape :", eps.shape)

# Cycles reprsentatifs
cycles_exemple = [0, 50, 200]

plt.figure(figsize=(12, 4))
for c in cycles_exemple:
    plt.plot(eps.iloc[c], label=f"Cycle {c}")

plt.title("EPS1 - Signal lectrique - Cycles exemple")
plt.xlabel("Temps")
plt.ylabel("EPS")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

eps_features = pd.DataFrame({
    "EPS1_mean": eps.mean(axis=1),
    "EPS1_median": eps.median(axis=1),
    "EPS1_std": eps.std(axis=1),
    "EPS1_max": eps.max(axis=1),
    "EPS1_rms": np.sqrt(np.mean(eps.values**2, axis=1))
})

print(eps_features.head())
print("Shape EPS1 features :", eps_features.shape)
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(
    [eps_features["EPS1_mean"], eps_features["EPS1_median"]],
    labels=["Mean", "Median"]
)
plt.title("EPS1 - Mean vs Median")
plt.ylabel("EPS")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(eps_features["EPS1_std"], bins=50)
plt.title("EPS1 - Distribution de l'cart-type")
plt.xlabel("STD")
plt.ylabel("Nombre de cycles")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(eps_features["EPS1_max"], bins=50)
plt.title("EPS1 - Distribution des maxima")
plt.xlabel("EPS max")
plt.ylabel("Nombre de cycles")
plt.grid(True)
plt.show()
```

```python
eps1_final = eps_features[[
    "EPS1_median",
    "EPS1_std",
    "EPS1_max"
]]
eps1_final.index.name = "cycle"
eps1_final.to_csv("dataset_EPS1.csv")

print("dataset_EPS1.csv sauvegard :", eps1_final.shape)

print(eps1_final.head())
print("Shape EPS1 final :", eps1_final.shape)
```
