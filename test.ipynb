```python
import pandas as pd

profile = pd.read_csv(
    "profile.txt",
    header=None,
    delim_whitespace=True
)

print(profile.shape)
print(profile.head())
```

```python
# colonne 1 = valve (100 = optimal)
profile["valve_score"] = profile.iloc[:, 1]

# le cycle = index (0 à 2204)
profile["cycle"] = profile.index
```

```python
cycles_non_optimaux = profile.loc[
    profile["valve_score"] < 100, "cycle"
]

print("Nombre de cycles NON optimaux :", len(cycles_non_optimaux))
print("Exemples :", cycles_non_optimaux.head(10).tolist())
```
