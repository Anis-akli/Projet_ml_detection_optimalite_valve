```python
import pandas as pd
import matplotlib.pyplot as plt

cp = pd.read_csv("CP.txt", sep="\t", header=None)[0]

plt.figure(figsize=(8,4))
plt.plot(cp)
plt.title("CP Pression par cycle")
plt.xlabel("Cycle")
plt.ylabel("Pression")
plt.grid(True)
plt.show()
```

```python
cp_delta = cp.diff()

plt.figure(figsize=(8,4))
plt.plot(cp_delta)
plt.title("CP Variation entre cycles")
plt.xlabel("Cycle")
plt.ylabel("DELTA Pression")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(cp, bins=50)
plt.title("CP Distribution")
plt.xlabel("Pression")
plt.ylabel("Nombre de cycles")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(cp_delta.dropna(), bins=50)
plt.title("CP Distribution des variations")
plt.xlabel("DELTA Pression")
plt.ylabel("Nombre de cycles")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np

cp = pd.read_csv("CP.txt", sep="\t", header=None)[0]

df_cp = pd.DataFrame({
    "CP": cp,
    "CP_delta": cp.diff(),
    
})

df_cp.iloc[0, df_cp.columns.get_loc("CP_delta")] = 0

print("Shape CP dataset :", df_cp.shape)
df_cp.index.name = "cycle"
df_cp.to_csv("dataset_CP.csv")

print("dataset_CP.csv sauvegard :", df_cp.shape)

df_cp.head()
```
