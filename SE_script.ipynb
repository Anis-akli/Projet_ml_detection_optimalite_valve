```python
import pandas as pd
import matplotlib.pyplot as plt

se = pd.read_csv("SE.txt", sep="\t", header=None)[0]

print("Nombre total de cycles :", len(se))
```

```python
plt.figure(figsize=(8,4))
plt.plot(se)
plt.title("SE – Énergie spécifique par cycle")
plt.xlabel("Cycle")
plt.ylabel("SE")
plt.grid(True)
plt.show()
```

```python
se_delta = se.diff()

plt.figure(figsize=(8,4))
plt.plot(se_delta)
plt.title("SE – Variation entre cycles")
plt.xlabel("Cycle")
plt.ylabel("Δ SE")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(se, bins=50)
plt.title("SE – Distribution")
plt.xlabel("SE")
plt.ylabel("Nombre de cycles")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(se_delta.dropna(), bins=50)
plt.title("SE – Distribution des variations")
plt.xlabel("Δ SE")
plt.ylabel("Nombre de cycles")
plt.grid(True)
plt.show()
```
