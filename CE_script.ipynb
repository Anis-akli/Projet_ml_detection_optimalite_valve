```python
import pandas as pd
import matplotlib.pyplot as plt

ce = pd.read_csv("CE.txt", sep="\t", header=None)[0]

plt.figure(figsize=(8,4))
plt.plot(ce, marker="o")
plt.title("CE Énergie cumulée par cycle")
plt.xlabel("Cycle")
plt.ylabel("Énergie")
plt.grid(True)
plt.show()
```

```python
ce_diff = ce.diff().fillna(0)

plt.figure(figsize=(8,4))
plt.plot(ce_diff)
plt.title("CE Variation d'énergie entre cycles")
plt.xlabel("Cycle")
plt.ylabel("DELTA Énergie")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(ce, bins=50)
plt.title("CE Distribution de l'énergie")
plt.xlabel("Énergie")
plt.ylabel("Nombre de cycles")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(ce_diff, bins=50)
plt.title("CE Distribution des variations d'énergie")
plt.xlabel("DELTA Énergie")
plt.ylabel("Nombre de cycles")
plt.show()
```

- La distribution de la variation d’énergie (CE_delta) est fortement concentrée autour de zéro, indiquant un fonctionnement stable du système.

 Cependant, la présence de valeurs extrêmes, bien que rares, révèle des événements énergétiques anormaux potentiellement liés à des défauts du système hydraulique. Ces variations constituent donc des indicateurs pertinents de dégradation.

```python
df_ce = pd.DataFrame({
    "CE_value": ce,
    "CE_diff": ce_diff
})
df_ce.index.name = "cycle"
df_ce.to_csv("dataset_CE.csv")

print("dataset_CE.csv sauvegardé :", df_ce.shape)

print(df_ce.head())
print("Shape CE :", df_ce.shape)
```
