```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cycles à afficher (toujours les mêmes pour comparer)
cycles_exemple = [0, 50, 200]
```

```python
fs_data = {}

for i in [1, 2]:
    fs = pd.read_csv(f"FS{i}.txt", sep="\t", header=None)
    fs_data[f"FS{i}"] = fs
    print(f"FS{i} shape :", fs.shape)
```

```python
for i in [1, 2]:
    fs = fs_data[f"FS{i}"]

    plt.figure(figsize=(12, 4))
    for c in cycles_exemple:
        plt.plot(fs.iloc[c], label=f"Cycle {c}")

    plt.title(f"FS{i} - Débit - Cycles exemple")
    plt.xlabel("Temps")
    plt.ylabel("Débit")
    plt.legend()
    plt.show()
```

```python
for i in [1, 2]:
    fs = fs_data[f"FS{i}"]

    fs_mean = fs.mean(axis=1)
    fs_median = fs.median(axis=1)

    plt.figure(figsize=(6, 4))
    plt.boxplot([fs_mean, fs_median], labels=["Mean", "Median"])
    plt.title(f"FS{i} - Mean vs Median")
    plt.ylabel("Débit")
    plt.show()
```

```python
for i in [1, 2]:
    fs = fs_data[f"FS{i}"]
    fs_std = fs.std(axis=1)

    plt.figure(figsize=(6, 4))
    plt.hist(fs_std, bins=40)
    plt.title(f"FS{i} - Distribution de l'écart-type")
    plt.xlabel("STD du débit")
    plt.ylabel("Nombre de cycles")
    plt.show()
```

```python
for i in [1, 2]:
    fs = fs_data[f"FS{i}"]
    fs_max = fs.max(axis=1)

    plt.figure(figsize=(6, 4))
    plt.hist(fs_max, bins=40)
    plt.title(f"FS{i} - Distribution des maxima")
    plt.xlabel("Débit max")
    plt.ylabel("Nombre de cycles")
    plt.show()
```

```python
import pandas as pd

# Chargement des fichiers FS
fs_data = {}

for i in [1, 2]:
    fs = pd.read_csv(f"FS{i}.txt", sep="\t", header=None)
    fs_data[f"FS{i}"] = fs
    print(f"FS{i} shape :", fs.shape)

# Extraction des features par cycle

fs_features = {}

for i in [1, 2]:
    fs = fs_data[f"FS{i}"]

    fs_features[f"FS{i}_median"] = fs.median(axis=1)
    fs_features[f"FS{i}_std"]    = fs.std(axis=1)
    fs_features[f"FS{i}_max"]    = fs.max(axis=1)


# Dataset final FS

df_fs = pd.DataFrame(fs_features)
df_fs.index.name = "cycle"
df_fs.to_csv("dataset_FS.csv")

print("dataset_FS.csv sauvegardé :", df_fs.shape)

print("\nDataset FS (FS1 + FS2) :")
print(df_fs.shape)
print(df_fs.head())
```
